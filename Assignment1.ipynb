# 5. Plotting 

```python
# plot histogram for each operator 
fig, axs= plt.subplots(df_new["Operator"].nunique(), figsize=(10,30))
for i, operator in enumerate(df_new["Operator"].unique()):
    axs[i].hist(df_new.loc[df_new["Operator"]== operator, "Delay"], bins=50)
    axs[i].title.set_text(operator)
    
```

# 6.Wrangling

```python
import datetime
# create a new dataset, a single journey as a rwo and totaljournytime, totalstop and total congestion via
df_new['Time'] = pd.to_datetime(df_new['Time'], unit='s')
df_new['Times'] = pd.to_datetime(df_new['Time'], unit='s')

summary_df = df_new.groupby("VehicleJID").agg({"Time": ["max"],"Times": ["min"], "AtStop":["sum"], "Congestion": ["sum"], "Delay": ["mean"]})
summary_df.columns=["MaxTime","MinTime" ,"TotalStop", "TotalCongestion", "AverageDelay"]
# find total time in hours for a single journey

summary_df["TotalTime"]= summary_df["MaxTime"] - summary_df["MinTime"] 
summary_df["TotalTime"]=summary_df.TotalTime / np.timedelta64(1, "h")
# new data set with total time, total stop, total congestion and average delay for a single journy
summary_df.loc[:,"TotalStop" : "TotalTime"]
```

# 7. Correlating


```python
# plot scatter to explore the corrlation between Time and average delay for a single journy

summary_df.plot.scatter("AverageDelay","TotalTime", s=5)
plt.xlim(-1000, 1000)
plt.ylim(0,2)
plt.title("Correlation between Average delay and Total time")
plt.xlabel("Average delay")
plt.ylabel("Total time")


summary_df.plot.scatter("TotalStop","TotalTime", s=5)
plt.ylim(0,2)
plt.xlim(0,80)
plt.title("Correlation between Total Stop and Total time")
plt.xlabel("Total Stop")
plt.ylabel("Total time")


summary_df.plot.scatter("TotalCongestion","TotalTime", s=5)
plt.ylim(0,2)
plt.xlim(0,10)
plt.title("Correlation between Total Congestion and Total time")
plt.xlabel("Total Congestion")
plt.ylabel("Total time")



```

As the scatter plot it looks like there is postive correlation between total time and total stop (total time of a single journy increase with increase in the number of stops)

for the rest which is (total time tota vs l congestion and average delay and total time) it look like no correlation between the variables.
